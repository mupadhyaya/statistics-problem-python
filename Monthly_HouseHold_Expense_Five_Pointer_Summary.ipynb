```python
import matplotlib.pyplot as plt

# Monthly household expenses for groceries
expenses = [220, 290, 320, 380, 430, 500, 520, 610, 650, 720, 750, 825, 850, 900, 950, 1010, 1050, 1130, 1150, 1280, 1290, 1300, 1330, 1410, 1430, 1525, 1550, 1630, 1650, 1705]

# (a) Five-number summary (minimum, first quartile, median, third quartile, maximum)
five_number_summary = (min(expenses), sorted(expenses)[int(len(expenses) / 4)], sorted(expenses)[int(len(expenses) / 2)], sorted(expenses)[int(3 * len(expenses) / 4)], max(expenses))

print("Five-number summary:", five_number_summary)

# (b) Box plot
plt.boxplot(expenses)
plt.xlabel("Month")
plt.ylabel("Grocery Expenses (USD)")
plt.title("Monthly Household Grocery Expenses")
plt.grid(True)
plt.show()

# Identifying potential outliers (IQR method)
Q1 = five_number_summary[1]
Q3 = five_number_summary[3]
IQR = Q3 - Q1

# Lower bound for outliers
lower_bound = Q1 - 1.5 * IQR
# Upper bound for outliers
upper_bound = Q3 + 1.5 * IQR

print("Lower bound:", lower_bound)
print("Upper bound:", upper_bound)

# Print potential outliers
potential_outliers = [expense for expense in expenses if expense < lower_bound or expense > upper_bound]
if potential_outliers:
  print("Potential outliers:", potential_outliers)
else:
  print("No potential outliers detected using IQR method.")
```
